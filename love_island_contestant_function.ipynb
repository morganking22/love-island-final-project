```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd

urls = [
    #Season 1
"https://loveisland.fandom.com/wiki/Jessica_Hayes",
"https://loveisland.fandom.com/wiki/Max_Morley",
"https://loveisland.fandom.com/wiki/Hannah_Elizabeth",
"https://loveisland.fandom.com/wiki/Jon_Clark",
"https://loveisland.fandom.com/wiki/Josh_Ritchie",
"https://loveisland.fandom.com/wiki/Lauren_Richardson",
"https://loveisland.fandom.com/wiki/Cally_Jane_Beech",
"https://loveisland.fandom.com/wiki/Luis_Morrison",
"https://loveisland.fandom.com/wiki/Jordan_Ring",
"https://loveisland.fandom.com/wiki/Zoe_Basia_Brown",
"https://loveisland.fandom.com/wiki/Ben_Porter",
"https://loveisland.fandom.com/wiki/Poppy_Farnan",
"https://loveisland.fandom.com/wiki/Naomi_Ball",
"https://loveisland.fandom.com/wiki/Travis_Almond",
"https://loveisland.fandom.com/wiki/Omar_Sultani",
"https://loveisland.fandom.com/wiki/Bethany_Rogers",
"https://loveisland.fandom.com/wiki/Daisy_Muller",
"https://loveisland.fandom.com/wiki/Chris_Williamson",
"https://loveisland.fandom.com/wiki/Chris_Baxter",
"https://loveisland.fandom.com/wiki/Danielle_Pyne",
"https://loveisland.fandom.com/wiki/Rachel_Christie",
"https://loveisland.fandom.com/wiki/John_Alberti",
"https://loveisland.fandom.com/wiki/Tony_Alberti",
# # #Season 2
"https://loveisland.fandom.com/wiki/Cara_De_La_Hoyde",
"https://loveisland.fandom.com/wiki/Nathan_Massey",
"https://loveisland.fandom.com/wiki/Alex_Bowen",
"https://loveisland.fandom.com/wiki/Olivia_Buckland",
"https://loveisland.fandom.com/wiki/Kady_McDermott",
"https://loveisland.fandom.com/wiki/Scott_Thomas",
"https://loveisland.fandom.com/wiki/Adam_Maxted",
"https://loveisland.fandom.com/wiki/Katie_Salmon",
"https://loveisland.fandom.com/wiki/Emma_Jane_Woodham",
"https://loveisland.fandom.com/wiki/Terry_Walsh",
"https://loveisland.fandom.com/wiki/Adam_Jukes",
"https://loveisland.fandom.com/wiki/Lauren_Whiteside",
"https://loveisland.fandom.com/wiki/Tina_Stinnes",
"https://loveisland.fandom.com/wiki/Sophie_Gradon",
"https://loveisland.fandom.com/wiki/Troy_Frith",
"https://loveisland.fandom.com/wiki/Tom_Powell",
"https://loveisland.fandom.com/wiki/Liana_Isadora_Van_Riel",
"https://loveisland.fandom.com/wiki/Oliver_Maxwell_Fernandez",
"https://loveisland.fandom.com/wiki/Malin_Andersson",
"https://loveisland.fandom.com/wiki/Zara_Holland",
"https://loveisland.fandom.com/wiki/James_Khan",
"https://loveisland.fandom.com/wiki/Rykard_Jenkins",
"https://loveisland.fandom.com/wiki/Daniel_Lukakis",
"https://loveisland.fandom.com/wiki/Rachel_Fenton",
"https://loveisland.fandom.com/wiki/Malia_Arkian",
"https://loveisland.fandom.com/wiki/Javi_Shephard",
# # #Season 3
"https://loveisland.fandom.com/wiki/Amber_Davies",
"https://loveisland.fandom.com/wiki/Kem_Cetinay",
"https://loveisland.fandom.com/wiki/Camilla_Thurlow",
"https://loveisland.fandom.com/wiki/Jamie_Jewitt",
"https://loveisland.fandom.com/wiki/Chris_Hughes",
"https://loveisland.fandom.com/wiki/Olivia_Attwood",
"https://loveisland.fandom.com/wiki/Gabby_Allen",
"https://loveisland.fandom.com/wiki/Marcel_Somerville",
"https://loveisland.fandom.com/wiki/Alex_Beattie",
"https://loveisland.fandom.com/wiki/Montana_Brown",
"https://loveisland.fandom.com/wiki/Georgia_Harrison",
"https://loveisland.fandom.com/wiki/Sam_Gowland",
"https://loveisland.fandom.com/wiki/Mike_Thalassitis",
"https://loveisland.fandom.com/wiki/Tyla_Carr",
"https://loveisland.fandom.com/wiki/Theo_Campbell",
"https://loveisland.fandom.com/wiki/Jonny_Mitchell",
"https://loveisland.fandom.com/wiki/Craig_Lawson",
"https://loveisland.fandom.com/wiki/Nathan_Joseph",
"https://loveisland.fandom.com/wiki/Danielle_Sellers",
"https://loveisland.fandom.com/wiki/Chyna_Ellis",
"https://loveisland.fandom.com/wiki/Amelia_Peters",
"https://loveisland.fandom.com/wiki/Dom_Lever",
"https://loveisland.fandom.com/wiki/Ellisha-Jade_White",
"https://loveisland.fandom.com/wiki/Marino_Katsouris",
"https://loveisland.fandom.com/wiki/Rob_Lipsett",
"https://loveisland.fandom.com/wiki/Shannen_McGrath",
"https://loveisland.fandom.com/wiki/Steve_Ball",
"https://loveisland.fandom.com/wiki/Simon_Searles",
"https://loveisland.fandom.com/wiki/ChloÃ«_Crowhurst",
"https://loveisland.fandom.com/wiki/Jessica_Shears",
"https://loveisland.fandom.com/wiki/Tyne-Lexy_Clarson",
"https://loveisland.fandom.com/wiki/Harley_Judge",
# # #Season 4
"https://loveisland.fandom.com/wiki/Dani_Dyer",
"https://loveisland.fandom.com/wiki/Jack_Fincham",
"https://loveisland.fandom.com/wiki/Laura_Anderson",
"https://loveisland.fandom.com/wiki/Paul_Knops",
"https://loveisland.fandom.com/wiki/Josh_Denzel",
"https://loveisland.fandom.com/wiki/Kazimir_Crossley",
"https://loveisland.fandom.com/wiki/Megan_Barton_Hanson",
"https://loveisland.fandom.com/wiki/Wes_Nelson",
"https://loveisland.fandom.com/wiki/Alex_George",
"https://loveisland.fandom.com/wiki/Alexandra_Cane",
"https://loveisland.fandom.com/wiki/Jack_Fowler",
"https://loveisland.fandom.com/wiki/Laura_Crane",
"https://loveisland.fandom.com/wiki/Josh_Mair",
"https://loveisland.fandom.com/wiki/Stephanie_Lam",
"https://loveisland.fandom.com/wiki/Georgia_Steel",
"https://loveisland.fandom.com/wiki/Sam_Bird",
"https://loveisland.fandom.com/wiki/Charlie_Brake",
"https://loveisland.fandom.com/wiki/Ellie_Brown",
"https://loveisland.fandom.com/wiki/Idris_Virgo",
"https://loveisland.fandom.com/wiki/Kieran_Nicholls",
"https://loveisland.fandom.com/wiki/Samira_Mighty",
"https://loveisland.fandom.com/wiki/Frankie_Foster",
"https://loveisland.fandom.com/wiki/Grace_Wardle",
"https://loveisland.fandom.com/wiki/Adam_Collard",
"https://loveisland.fandom.com/wiki/Alex_Miller",
"https://loveisland.fandom.com/wiki/Darylle_Sargeant",
"https://loveisland.fandom.com/wiki/Ellie_Jones",
"https://loveisland.fandom.com/wiki/Charlie_Williams",
"https://loveisland.fandom.com/wiki/Dean_Overson",
"https://loveisland.fandom.com/wiki/Jordan_Adefeyisan",
"https://loveisland.fandom.com/wiki/Savanna_Darnell",
"https://loveisland.fandom.com/wiki/Zara_McDermott",
"https://loveisland.fandom.com/wiki/Eyal_Booker",
"https://loveisland.fandom.com/wiki/Rosie_Williams",
"https://loveisland.fandom.com/wiki/Charlie_Frederick",
"https://loveisland.fandom.com/wiki/Hayley_Hughes",
"https://loveisland.fandom.com/wiki/Niall_Aslam",
"https://loveisland.fandom.com/wiki/Kendall_Rae-Knight",
# # #Season 5
"https://loveisland.fandom.com/wiki/Amber_Gill",
"https://loveisland.fandom.com/wiki/Greg_O%27Shea",
"https://loveisland.fandom.com/wiki/Molly-Mae_Hague",
"https://loveisland.fandom.com/wiki/Tommy_Fury",
"https://loveisland.fandom.com/wiki/India_Reynolds",
"https://loveisland.fandom.com/wiki/Ovie_Soko",
"https://loveisland.fandom.com/wiki/Curtis_Pritchard",
"https://loveisland.fandom.com/wiki/Maura_Higgins",
"https://loveisland.fandom.com/wiki/Anton_Danyluk",
"https://loveisland.fandom.com/wiki/Belle_Hassan",
"https://loveisland.fandom.com/wiki/Anna_Vakili",
"https://loveisland.fandom.com/wiki/Chris_Taylor",
"https://loveisland.fandom.com/wiki/Harley_Brash",
"https://loveisland.fandom.com/wiki/Jordan_Hames",
"https://loveisland.fandom.com/wiki/Francesca_Allen",
"https://loveisland.fandom.com/wiki/Michael_Griffiths",
"https://loveisland.fandom.com/wiki/Joanna_Chimonides",
"https://loveisland.fandom.com/wiki/Marvin_Brooks",
"https://loveisland.fandom.com/wiki/Amy_Hart",
"https://loveisland.fandom.com/wiki/George_Rains",
"https://loveisland.fandom.com/wiki/Lucie_Donlan",
"https://loveisland.fandom.com/wiki/Danny_Williams",
"https://loveisland.fandom.com/wiki/Jourdan_Riane",
"https://loveisland.fandom.com/wiki/Dan_Rose",
"https://loveisland.fandom.com/wiki/Dennon_Lewis",
"https://loveisland.fandom.com/wiki/Lavena_Back",
"https://loveisland.fandom.com/wiki/Maria_Wild",
"https://loveisland.fandom.com/wiki/Nabila_Badda",
"https://loveisland.fandom.com/wiki/Stevie_Bradley",
"https://loveisland.fandom.com/wiki/Arabella_Chi",
"https://loveisland.fandom.com/wiki/Tom_Walker",
"https://loveisland.fandom.com/wiki/Yewande_Biala",
"https://loveisland.fandom.com/wiki/Elma_Pazar",
"https://loveisland.fandom.com/wiki/Joe_Garratt",
"https://loveisland.fandom.com/wiki/Sherif_Lanre",
"https://loveisland.fandom.com/wiki/Callum_Macleod",
# # #Season 6
"https://loveisland.fandom.com/wiki/Finley_Tapp",
"https://loveisland.fandom.com/wiki/Paige_Turley",
"https://loveisland.fandom.com/wiki/Luke_Trotman",
 "https://loveisland.fandom.com/wiki/Siannise_Fudge",
"https://loveisland.fandom.com/wiki/Demi_Jones",
"https://loveisland.fandom.com/wiki/Luke_Mabbott",
"https://loveisland.fandom.com/wiki/Ched_Uzor",
"https://loveisland.fandom.com/wiki/Jess_Gale",
"https://loveisland.fandom.com/wiki/Mike_Boateng",
"https://loveisland.fandom.com/wiki/Priscilla_Anyabu",
"https://loveisland.fandom.com/wiki/Callum_Jones",
"https://loveisland.fandom.com/wiki/Molly_Smith",
"https://loveisland.fandom.com/wiki/Jamie_Clayton",
"https://loveisland.fandom.com/wiki/Natalia_Zoppa",
"https://loveisland.fandom.com/wiki/Shaughna_Phillips",
"https://loveisland.fandom.com/wiki/Eva_Zapico",
"https://loveisland.fandom.com/wiki/Jordan_Waobikeze",
"https://loveisland.fandom.com/wiki/Nas_Majeed",
"https://loveisland.fandom.com/wiki/Rebecca_Gormley",
"https://loveisland.fandom.com/wiki/Alexi_Eraclides",
"https://loveisland.fandom.com/wiki/Biggs_Chris",
"https://loveisland.fandom.com/wiki/George_Day",
"https://loveisland.fandom.com/wiki/Jade_Affleck",
"https://loveisland.fandom.com/wiki/Jamie_McCann",
"https://loveisland.fandom.com/wiki/Josh_Kempton",
"https://loveisland.fandom.com/wiki/Sophie_Piper",
"https://loveisland.fandom.com/wiki/Wallace_Wilson",
"https://loveisland.fandom.com/wiki/Leanne_Amaning",
"https://loveisland.fandom.com/wiki/Connor_Durman",
"https://loveisland.fandom.com/wiki/Connagh_Howard",
"https://loveisland.fandom.com/wiki/Eve_Gale",
"https://loveisland.fandom.com/wiki/Ollie_Williams",
# # #Season 7
"https://loveisland.fandom.com/wiki/Liam_Reardon",
"https://loveisland.fandom.com/wiki/Millie_Court",
"https://loveisland.fandom.com/wiki/Chloe_Burrows",
"https://loveisland.fandom.com/wiki/Toby_Aromolaran",
"https://loveisland.fandom.com/wiki/Faye_Winter",
"https://loveisland.fandom.com/wiki/Teddy_Soares",
"https://loveisland.fandom.com/wiki/Kaz_Kamwi",
"https://loveisland.fandom.com/wiki/Tyler_Cruickshank",
"https://loveisland.fandom.com/wiki/Jake_Cornish",
"https://loveisland.fandom.com/wiki/Liberty_Poole",
"https://loveisland.fandom.com/wiki/Aaron_Simpson",
"https://loveisland.fandom.com/wiki/Mary_Bedford",
"https://loveisland.fandom.com/wiki/Brett_Staniland",
"https://loveisland.fandom.com/wiki/Priya_Gopaldas",
"https://loveisland.fandom.com/wiki/Matthew_MacNabb",
"https://loveisland.fandom.com/wiki/Abigail_Rawlings",
"https://loveisland.fandom.com/wiki/Dale_Mehmet",
"https://loveisland.fandom.com/wiki/Amy_Day",
"https://loveisland.fandom.com/wiki/Clarisse_Juliette",
"https://loveisland.fandom.com/wiki/Hugo_Hammond",
"https://loveisland.fandom.com/wiki/Sam_Jackson",
"https://loveisland.fandom.com/wiki/Harry_Young",
"https://loveisland.fandom.com/wiki/Jack_Barlow",
"https://loveisland.fandom.com/wiki/Kaila_Troy",
"https://loveisland.fandom.com/wiki/Lillie_Haynes",
"https://loveisland.fandom.com/wiki/Medhy_Malanda",
"https://loveisland.fandom.com/wiki/Salma_Naran",
"https://loveisland.fandom.com/wiki/Aaron_Francis",
"https://loveisland.fandom.com/wiki/Lucinda_Strafford",
"https://loveisland.fandom.com/wiki/Georgia_Townend",
"https://loveisland.fandom.com/wiki/Andrea-Jane_Bunker",
"https://loveisland.fandom.com/wiki/Danny_Bibby",
"https://loveisland.fandom.com/wiki/Sharon_Gaffka",
"https://loveisland.fandom.com/wiki/Brad_McClelland",
"https://loveisland.fandom.com/wiki/Rachel_Finni",
"https://loveisland.fandom.com/wiki/Chuggs_Wallis",
"https://loveisland.fandom.com/wiki/Shannon_Singh",
# # #Season 8
"https://loveisland.fandom.com/wiki/Davide_Sanclimenti",
"https://loveisland.fandom.com/wiki/Ekin-Su_CÃ¼lcÃ¼loÄlu",
"https://loveisland.fandom.com/wiki/Gemma_Owen",
"https://loveisland.fandom.com/wiki/Luca_Bish",
"https://loveisland.fandom.com/wiki/Dami_Hope",
"https://loveisland.fandom.com/wiki/Indiyah_Polack",
"https://loveisland.fandom.com/wiki/Andrew_Le_Page",
"https://loveisland.fandom.com/wiki/Tasha_Ghouri",
"https://loveisland.fandom.com/wiki/Adam_Collard",
"https://loveisland.fandom.com/wiki/Paige_Thorne",
"https://loveisland.fandom.com/wiki/Danica_Taylor",
"https://loveisland.fandom.com/wiki/Jamie_Allen",
"https://loveisland.fandom.com/wiki/Deji_Adeniyi",
"https://loveisland.fandom.com/wiki/Lacey_Edwards",
"https://loveisland.fandom.com/wiki/Nathalia_Campos",
"https://loveisland.fandom.com/wiki/Reece_Ford",
"https://loveisland.fandom.com/wiki/Billy_Brown",
"https://loveisland.fandom.com/wiki/Summer_Botwe",
"https://loveisland.fandom.com/wiki/Coco_Lodge",
"https://loveisland.fandom.com/wiki/Josh_Le_Grove",
"https://loveisland.fandom.com/wiki/Jacques_O%27Neill",
"https://loveisland.fandom.com/wiki/Chyna_Mills",
"https://loveisland.fandom.com/wiki/Jay_Younger",
"https://loveisland.fandom.com/wiki/Cheyanne_Kerr",
"https://loveisland.fandom.com/wiki/George_Tasker",
"https://loveisland.fandom.com/wiki/Jack_Keating",
"https://loveisland.fandom.com/wiki/Jazmine_Nichol",
"https://loveisland.fandom.com/wiki/Mollie_Salmon",
"https://loveisland.fandom.com/wiki/Samuel_Agbiji",
"https://loveisland.fandom.com/wiki/Antigoni_Buxton",
"https://loveisland.fandom.com/wiki/Charlie_Radnedge",
"https://loveisland.fandom.com/wiki/Amber_Beckford",
"https://loveisland.fandom.com/wiki/Ikenna_Ekwonna",
"https://loveisland.fandom.com/wiki/Remi_Lambert",
"https://loveisland.fandom.com/wiki/Afia_Tonkmor",
"https://loveisland.fandom.com/wiki/Liam_Llewellyn",
# # #Season 9
"https://loveisland.fandom.com/wiki/Kai_Fagan",
"https://loveisland.fandom.com/wiki/Sanam_Harrinanan",
"https://loveisland.fandom.com/wiki/Lana_Jenkins",
"https://loveisland.fandom.com/wiki/Ron_Hall",
"https://loveisland.fandom.com/wiki/Samie_Elishi",
"https://loveisland.fandom.com/wiki/Tom_Clare",
"https://loveisland.fandom.com/wiki/Shaq_Muhammad",
"https://loveisland.fandom.com/wiki/Tanya_Manhenga",
"https://loveisland.fandom.com/wiki/Jessie_Wynter",
"https://loveisland.fandom.com/wiki/Will_Young",
"https://loveisland.fandom.com/wiki/Casey_O%27Gorman",
"https://loveisland.fandom.com/wiki/Rosie_Seabrook",
"https://loveisland.fandom.com/wiki/Claudia_Fogarty",
"https://loveisland.fandom.com/wiki/Keanan_Brand",
"https://loveisland.fandom.com/wiki/Maxwell_Samuda",
"https://loveisland.fandom.com/wiki/Olivia_Hawkins",
"https://loveisland.fandom.com/wiki/Martin_Akinola",
"https://loveisland.fandom.com/wiki/Bayley_Mummery",
"https://loveisland.fandom.com/wiki/Cynthia_Otseh-Taiwo",
"https://loveisland.fandom.com/wiki/Frankie_Davey",
"https://loveisland.fandom.com/wiki/Kain_Reed",
"https://loveisland.fandom.com/wiki/Layla_Al-Momani",
"https://loveisland.fandom.com/wiki/Lydia_Karakyriakou",
"https://loveisland.fandom.com/wiki/Lynda_Flix",
"https://loveisland.fandom.com/wiki/Ryan_Weekley",
"https://loveisland.fandom.com/wiki/Sammy_James",
"https://loveisland.fandom.com/wiki/Ellie_Spence",
"https://loveisland.fandom.com/wiki/Jordan_Odofin",
"https://loveisland.fandom.com/wiki/Tanyel_Revan",
"https://loveisland.fandom.com/wiki/Aaron_Waters",
"https://loveisland.fandom.com/wiki/Spencer_Wilks",
"https://loveisland.fandom.com/wiki/Zara_Lackenby-Brown",
"https://loveisland.fandom.com/wiki/Anna-May_Robey",
"https://loveisland.fandom.com/wiki/Haris_Namani",
"https://loveisland.fandom.com/wiki/David_Salako",
# # #Season 10
"https://loveisland.fandom.com/wiki/Jess_Harding",
"https://loveisland.fandom.com/wiki/Sammy_Root",
"https://loveisland.fandom.com/wiki/Lochan_Nowacki",
"https://loveisland.fandom.com/wiki/Whitney_Adebayo",
"https://loveisland.fandom.com/wiki/Ella_Thomas",
"https://loveisland.fandom.com/wiki/Tyrique_Hyde",
"https://loveisland.fandom.com/wiki/Molly_Marsh",
"https://loveisland.fandom.com/wiki/Zachariah_Noble",
"https://loveisland.fandom.com/wiki/Ella_Barnes",
"https://loveisland.fandom.com/wiki/Mitchel_Taylor",
"https://loveisland.fandom.com/wiki/Abi_Moores",
"https://loveisland.fandom.com/wiki/Scott_van-der-Sluis",
"https://loveisland.fandom.com/wiki/Amber_Wise",
"https://loveisland.fandom.com/wiki/Josh_Brocklebank",
"https://loveisland.fandom.com/wiki/Kady_McDermott",
"https://loveisland.fandom.com/wiki/Ouzy_See",
"https://loveisland.fandom.com/wiki/Catherine_Agbaje",
"https://loveisland.fandom.com/wiki/Elom_Ahlijah-Wilson",
"https://loveisland.fandom.com/wiki/Leah_Taylor",
"https://loveisland.fandom.com/wiki/Montel_McKenzie",
"https://loveisland.fandom.com/wiki/Benjamin_Noel",
"https://loveisland.fandom.com/wiki/Danielle_Mazhindu",
"https://loveisland.fandom.com/wiki/Gabby_Jeffery",
"https://loveisland.fandom.com/wiki/Kodie_Murphy",
"https://loveisland.fandom.com/wiki/Tink_Reading",
"https://loveisland.fandom.com/wiki/Zachary_Ashford",
"https://loveisland.fandom.com/wiki/Malissa_Nicol",
"https://loveisland.fandom.com/wiki/Mehdi_Edno",
"https://loveisland.fandom.com/wiki/AndrÃ©_Furtado",
"https://loveisland.fandom.com/wiki/Charlotte_Sumner",
"https://loveisland.fandom.com/wiki/Ruchee_Gurung",
"https://loveisland.fandom.com/wiki/George_Fensom"
]


def build_contestant(soup):

    name = soup.find("h2", {"data-source": "Name"}).get_text()
    image_href = soup.find("a", {"class": "image-thumbnail"})['href']
    entered = soup.find("td", {"data-source": "Entered"}).get_text()    

    left = soup.find("td", {"data-source": "Left"})
    if left:
        left = left.get_text()
    
    gender = soup.find("div", {"data-source": "Gender"})
    if gender:
        gender = gender.findNext('div').get_text()

    age = soup.find("div", {"data-source": "Age"})
    if age:
        age = age.findNext('div').get_text()

    birthdate = soup.find("div", {"data-source": "Born"})
    if birthdate:
        birthdate = birthdate.findNext('div').get_text()

    occupation = soup.find("div", {"data-source": "Occupation"})
    if occupation:
        occupation = occupation.findNext('div').get_text()

    hometown = soup.find("div", {"data-source": "Hometown"})
    if hometown:
        hometown = hometown.findNext('div').get_text()
    
    spouses = soup.find("div", {"data-source": "Spouses"})
    if spouses:
        spouses = spouses.findNext('div').get_text()

    children = soup.find("div", {"data-source": "Children"})
    if children:
        children = children.findNext('div').get_text()

    status = soup.find("div", {"data-source": "Status"})
    if status:
        status = status.findNext('div').get_text()

    contestant = {
        'name': name,
        'image_href': image_href,
        'gender': gender,
        'age': age,
        'birthdate': birthdate,
        'occupation': occupation,
        'hometown': hometown,
        'spouses': spouses,
        'children': children,
        'entered': entered,
        'left': left,
        'status': status,
        'spouses': spouses,
    }

    return contestant


all_contestants = []

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    contestant = build_contestant(soup)
    all_contestants.append(contestant)

print(all_contestants)

df = pd.DataFrame(all_contestants)
print(df)
df.to_csv('love_island_export.csv')
```
