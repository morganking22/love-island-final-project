```python
from bs4 import BeautifulSoup
import requests
import csv
#print(response.status_code)

url = "https://loveisland.fandom.com/wiki/Love_Island_(Season_1)"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

tables = soup.find_all("table", {"class": "wikitable"})
islanders = tables[0]

rows = islanders.findAll("tr")

contenstant_data = []

for row in rows:
    tds = row.find_all("td")

    for count, td in enumerate(tds):
        contestant = {}
        text = td.get_text()
        text = text.strip('\n')

        match count:
            case 0:
                #hex color
                pass
            case 1:
                #name and href
                contestant.update({'name': text})
            case 2:
                #age
                contestant.update({'age': text})
            case 3:
                #hometown
                contestant.update({'hometown': text})
            case 4:
                #date entered
                contestant.update({'date_entered': text})
            case 5:
                #status on the island
                if text:
                    contestant.update({'status': text})

        if bool(contestant):
            contenstant_data.append(contestant)

        print(contestant)

contestant.to_csv("love_island_season_1_scrape.csv")
```
